```
import cv2
import numpy as np 
from matplotlib import pyplot as plt
%matplotlib inline

image = cv2.imread('cursoDL/imagens/car.png',0)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1, xticks=[], yticks=[])
ax.imshow(image,cmap='gray')
plt.title('Imagem original')

```

```
filtrada = np.zeros((image.shape[0], image.shape[1]))


#filtro media
for linha in range(2, image.shape[0]-2):
  for col in range(2, image.shape[1]-2):
    filtrada[linha,col]= (np.sum(image[linha-2:linha+3, col-2:col+3]))/25

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1, xticks=[], yticks=[])
ax.imshow(filtrada,cmap='gray')
plt.title('Imagem filtrada')    
    
```
