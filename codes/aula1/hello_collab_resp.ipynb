```
import cv2
from matplotlib import pyplot as plt

# Plot inline
%matplotlib inline

#lê as informações básicas da imagem
im = cv2.imread("cursoDL/imagens/thanos.jpg")
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

print("largura: %d pixels"%im.shape[1])
print("altura: %d pixels"%im.shape[0])

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1, xticks=[], yticks=[])
ax.imshow(im)
plt.title('Primeira Figura')
```
